```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("/content/Sport car price (1).csv")

df["Price (in USD)"] = df["Price (in USD)"].replace(r"[\$,]", "", regex=True).astype(float)

fig_treemap = px.treemap(
    df,
    path=["Car Make", "Car Model", "Year"],
    values="Price (in USD)",
    color="Horsepower",
    color_continuous_scale="Viridis",
    title="Treemap of Car Make → Car Model → Year"
)
fig_treemap.show()

fig_sunburst = px.sunburst(
    df,
    path=["Car Make", "Car Model", "Year"],
    values="Price (in USD)",
    color="Horsepower",
    color_continuous_scale="Viridis",
    title="Sunburst of Car Make → Car Model → Year"
)
fig_sunburst.show()
x
```

```python
!pip install kagglehub plotly pandas --quiet

import kagglehub
import pandas as pd
import plotly.express as px
import os

path = kagglehub.dataset_download("rkiattisak/sports-car-prices-dataset")

csv_files = [file for file in os.listdir(path) if file.endswith('.csv')]

df = pd.read_csv(os.path.join(path, csv_files[0]))

df["Price (in USD)"] = df["Price (in USD)"].replace(r"[\$,]", "", regex=True).astype(float)

fig = px.treemap(df, path=['Car Make', 'Car Model'], values='Price (in USD)', color='Price (in USD)', color_continuous_scale='Viridis', title='Treemap of Sports Car Prices by Brand and Model')
fig.show()

fig = px.sunburst(df, path=['Car Make', 'Car Model'], values='Price (in USD)', color='Price (in USD)', color_continuous_scale='RdBu', title='Sunburst of Sports Car Prices by Brand and Model')
fig.show()
```
