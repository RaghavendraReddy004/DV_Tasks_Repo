```python
!pip install seaborn matplotlib pandas joypy

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from joypy import joyplot

df = pd.read_csv("/content/Sport car price.csv")

categorical_cols = df.select_dtypes(include=['object']).columns
numerical_cols = df.select_dtypes(include=['int64','float64']).columns

cat = categorical_cols[0]   # pick first categorical column
num = numerical_cols[0]     # pick first numerical column

# 1. Bar Chart (mean value of num grouped by cat)
sns.barplot(x=cat, y=num, data=df, estimator="mean", ci=None)
plt.title(f"Bar Chart - {cat} vs {num}")
plt.xticks(rotation=45)
plt.show()

# 2. Grouped Kernel Density Plot
sns.kdeplot(data=df, x=num, hue=cat, fill=True)
plt.title(f"Grouped KDE Plot - {cat} vs {num}")
plt.show()

# 3. Box Plot
sns.boxplot(x=cat, y=num, data=df)
plt.title(f"Box Plot - {cat} vs {num}")
plt.xticks(rotation=45)
plt.show()

# 4. Violin Plot
sns.violinplot(x=cat, y=num, data=df)
plt.title(f"Violin Plot - {cat} vs {num}")
plt.xticks(rotation=45)
plt.show()

# 5. Ridgeline Plot
plt.figure(figsize=(10,6))
joyplot(data=df, by=cat, column=num, figsize=(10,6))
plt.title(f"Ridgeline Plot - {cat} vs {num}")
plt.show()

# 6. Beeswarm Plot
sns.swarmplot(x=cat, y=num, data=df)
plt.title(f"Beeswarm Plot - {cat} vs {num}")
plt.xticks(rotation=45)
plt.show()

# 7. Scatter Plot (first two numerical variables)
if len(numerical_cols) >= 2:
    sns.scatterplot(x=df[numerical_cols[0]], y=df[numerical_cols[1]], hue=df[cat])
    plt.title(f"Scatter Plot - {numerical_cols[0]} vs {numerical_cols[1]} by {cat}")
    plt.show()
```
